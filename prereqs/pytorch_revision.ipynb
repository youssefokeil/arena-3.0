```python
import torch
```

## Tensors

Why use Tensors over Numpy's ndarrays?

Answer: Tensors are 
1. can run on GPUs, meaning operations can be parallelized
2. optimized for automatic differentiation, making it extra fast for neural networks and gradient descent.

### Initializing a Tensor

#### tensors can be created directly from data.

```python
data = [[0,1], [2,3]]
x_data = torch.tensor(data)
```

#### made from numpy array

```python
import numpy as np
```

```python
rng = np.random.default_rng(47)
data2 = rng.random((3,3,3))
x_data2 = torch.from_numpy(data2)
x_data2
```

#### made from another tensor

```python
x_ones = torch.ones_like(x_data2)
print(f"The ones tensors: \n {x_ones} \n")

x_rand = torch.rand_like(x_data2)
print(f"Random tensor: \n {x_rand} \n")

x_zeros = torch.zeros_like(x_data2)
print(f"Zeros tensor: \n {x_zeros} \n")
```

### Tensor Attributes

to know the properties/attributes of a tensor there are a couple of pytorch helpful methods

```python
tensor =  torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Number of dimensions of tensor: {tensor.ndim}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")

## you can change device tensor is stored on using this

tensor = tensor.to(device=torch.device('cuda'))

print(f"New device for tensor is: {tensor.device}")
```

### Operations on Tensors

there are a lot of operations on tensors, which we don't have to revise all of them now

#### Joining Tensors

concatenate sequence of tensors along a dimension

```python
tensor2 = torch.rand(1,2)

tensor2
```

```python
## horizontal stacking
tensor2_horizontal = torch.cat([tensor2, tensor2, tensor2], dim=1)
print(tensor2_horizontal)
```

```python
## vertical stacking
tensor2_vertical = torch.cat([tensor2, tensor2, tensor2], dim=0)
print(tensor2_vertical)
```

#### Arithmetic

some arithmetic operations include matrix multiplication usin `@`, `matmul`

```python
tensor_np = rng.integers(10, size=(3,3))
tensor1 = torch.from_numpy(tensor_np)
tensor2 = torch.from_numpy(rng.integers(10, size=(3,2)))
print(f"Tensor1: \n {tensor1}")
print(f"Tensor2: \n {tensor2}")
```

```python
## the matrix multiplication
y1 = tensor1 @ tensor2
y2 = tensor1.matmul(tensor2)
print(f"y1: \n {y1}")
print(f"y2: \n {y2}")
```

another one is element-wise multiplication

```python
y3 = tensor1 * tensor1
y4 = tensor1.mul(tensor1)
print(f"y3: \n {y3}")
print(f"y4: \n {y4}")
```
